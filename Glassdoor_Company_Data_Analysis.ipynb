````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
````

1. Load the data


````python
df = pd.read_csv("glassdoor_mock_data.csv")
df
````

2. Avergae Salary

````python
avg_salary = df['Salary_Estimate'].mean()
fig, ax = plt.subplots(figsize=(6, 5))
ax.set_xlim(4, 10)
ax.set_ylim(0, 5)
ax.axis('off')
ax.text(4.8, 4, 'Average Salary', fontsize=28, weight='bold')
ax.text(5.5, 3, f'₹{avg_salary:,.0f}', fontsize=30, color='green',weight='bold')
plt.tight_layout()
plt.show()
````

3. Average Rating

````python
avg_rating = df['Rating'].mean()
fig, ax = plt.subplots(figsize=(6, 5))
ax.set_xlim(3, 10)
ax.set_ylim(0, 5)
ax.axis('off')
ax.text(4.2, 4, 'Average Rating', fontsize=25, weight='bold')
ax.text(6, 3, f'{avg_rating:,.0f}', fontsize=30, color='green',weight='bold')
plt.tight_layout()
plt.show()
````

4. Clean Columns Name

````python
df.columns = df.columns.str.strip().str.replace(" ", "_")
````

````python
df.head(5)
````

````python
df.columns
````

````python
df.info()
````

5. Convert Date Posted to Date TIme

````python
df['Date_Posted'] = pd.to_datetime(df['Date_Posted'], errors='coerce')
df.head(5)
````

6. Add 'Month_Posted' column

````python
df['Month_Posted'] = df['Date_Posted'].dt.to_period('M')
df.head(5)
````

7. Average Salary by Job Title

````python
avg_salary = df.groupby('Job_Title')['Salary_Estimate'].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 5))
avg_salary.plot(kind='bar', color='green')
plt.title('Top 5 Average Salaries by Job Title')
plt.ylabel('Average Salary')
plt.xlabel('Job Title')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

8. Remote Option Distribution

````python
remote_counts = df['Remote_Option'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(remote_counts, labels=remote_counts.index, autopct='%1.1f%%', startangle=90, colors=['green', 'maroon'])
plt.title('Remote Option Distribution')
plt.show()
````

9. Job Posting Overtime

````python
jobs_per_month = df['Month_Posted'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
jobs_per_month.plot(kind='line', color='green',marker= 'o')
plt.title('Job Postings Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Postings')
plt.grid(False)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

10. Histogram of Salary


````python
plt.figure(figsize=(10, 6))
plt.hist(df['Salary_Estimate'], bins=40, color='green', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()
````

11. Scatter Plot: Salary vs Rating

````python
plt.figure(figsize=(10, 6))
plt.scatter(df['Rating'], df['Salary_Estimate'], alpha=0.5, color='green')
plt.title('Salary vs. Company Rating')
plt.xlabel('Rating')
plt.ylabel('Salary')
plt.grid(False)
plt.tight_layout()
plt.show()
````
